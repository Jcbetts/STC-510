```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
city_list = ['Phoenix',
            'Scottsdale',
            'New York',
            'Buffalo',
            'Seattle',
            'La Jolla, CA',
            'Irvine',
            'Boulder',
            'Eugene',
            'Baltimore']
```

```python
city_list
```

```python
my_cities = gpd.tools.geocode(city_list)
```

```python
my_cities
```

```python
my_cities.plot()
```

```python
usa = gpd.read_file('cb_2018_us_state_500k.shp')
```

```python
usa.head
```

```python
usa.plot()
```

```python
usa.plot(figsize=(20,20))
```

```python
ax = usa[~usa.STUSPS.isin(['AK', 'HI', 'AS', 'PR', 'VI', 'GU', 'MP'])].plot(figsize=(20, 20), color='grey')
my_cities.plot(ax=ax, color='red')
plt.show()
```
