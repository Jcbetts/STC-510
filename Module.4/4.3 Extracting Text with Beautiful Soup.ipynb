```python
import requests
from bs4 import BeautifulSoup
```

```python
thispage = requests.get('https://www.phoenixnewtimes.com/restaurants/mesa-geogaching-tour-east-valley-farms-and-restaurants-11520225')
```

```python
thispage.text
```

```python
bs = BeautifulSoup(thispage.text,'html.parser')
```

```python
bs
```

```python
story_body_true = bs.find(class_='fdn-content-body').text
```

```python
print(story_body_true)
```

```python
story_body = bs.find(class_='ZoneA').text
```

```python
print(story_body)
```

```python
allps = story_body.find('p')
```

```python
alltext = ''
for eachp in allps:
    alltext += '\n\n' + eachp.text
```

```python
'\n\n'.join(allps.text)
```
